```python
%%writefile ordenamiento.cpp
#include<iostream>
#include<vector>
#include<time.h>
using namespace std;
const int n = 6;
void print_vec(vector<int>numeros)
{
	for (int numero : numeros)
	{
		cout << numero << " ";
	}
	cout << endl;
}
struct busqueda {
private:
	static bool _binaria(vector<int>numeros, int numero, int inicio, int fin)
	{
		if (fin >= inicio)
		{
			int mid = (fin + inicio) / 2;
			if (numeros[mid] == numero)
			{
				return true;
			}
			else if (numeros[mid] < numero)
			{
				return _binaria(numeros, numero, mid + 1, fin);
			}
			else
				return _binaria(numeros, numero, inicio, mid - 1);
		}
		return false;
	}
public:
	static bool lineal(vector<int>numeros, int numero)
	{
		//El mejor de los casos: 1
		//El peor de los casos: n
		//Complejidad: O(n)
		for (int i = 0; i < numeros.size(); i++)
		{
			if (numeros[i] == numero)
				return true;
		}
		return false;
	}
	static bool binaria(vector<int>numeros, int numero)
	{
		return _binaria(numeros, numero, 0, numeros.size() - 1);
	}
};
struct ordenamiento {
private:
	static void ordenar(vector<int>&numeros, int izquierda, int derecha, int medio)
	{
		
		vector<int>iz;
		vector<int>der;
		for (int i = izquierda; i <= medio; i++)
		{
			iz.push_back(numeros[i]);
		}
		for (int i = medio+1; i <= derecha; i++)
		{
			der.push_back(numeros[i]);
		}
		int i, j, v;
		i = j = 0;
		v = izquierda;
		while (i < iz.size() && j < der.size())
		{
			if (iz[i] <= der[j]) {
				numeros[v] = iz[i];
				i++;
				v++;
			}
			else {
				numeros[v] = der[j];
				j++;
				v++;
			}
		}
		while (i < iz.size())
		{
			numeros[v] = iz[i];
			i++;
			v++;
		}
		while (j < der.size())
		{
			numeros[v] = der[j];
			j++;
			v++;
		}
	}
	static void _merge(vector<int>& numeros, int izquierda, int derecha)
	{
		if (izquierda < derecha)
		{
			int medio = (izquierda + derecha) / 2;
			_merge(numeros, izquierda, medio);
			_merge(numeros, medio + 1, derecha);
			ordenar(numeros, izquierda, derecha, medio);
		}
	}
public:
	static void burbuja(vector<int> & numeros)
	{
		//El mejor: n
		//El peor: n^2
		bool ordenado;
		for (int i = 0; i < numeros.size()-1; i++)
		{
			ordenado = true;
			for (int j =0; j < numeros.size()-i-1; j++)
			{
				if (numeros[j] > numeros[j+1])
				{
					int aux = numeros[j];
					numeros[j] = numeros[j + 1];
					numeros[j + 1] = aux;
					ordenado = false;
				}
			}
			if (ordenado)break;
		}
	}
	static void seleccion(vector<int>&numeros)
	{
		for (int i = 0; i < numeros.size(); i++)
		{
			int menor = i;
			for (int j = i+1; j < numeros.size(); j++)
			{
				if (numeros[j] < numeros[menor])
				{
					menor = j;
				}
			}
			int aux = numeros[i];
			numeros[i] = numeros[menor];
			numeros[menor] = aux;
		}
	}
	static void intercambio(vector<int>&numeros)
	{
		bool ordenado = true;
		for (int i = 0; i < numeros.size()-1; i++)
		{
			if (numeros[i] > numeros[i + 1])
				ordenado = false;
		}
		if (ordenado) return;
		for (int i = 0; i < numeros.size() - 1; i++)
		{
			for (int j = i + 1; j < numeros.size(); j++)
			{
				if (numeros[i] > numeros[j])
				{
					int aux = numeros[i];
					numeros[i] = numeros[j];
					numeros[j] = aux;
				}
			}
		}
	}
	static void insercion(vector<int>& numeros)
	{
		//Mejor de los casos: n
		//Peor de los casos: n^2
		//Total: O(n^2);
		int cont = 0;
		int j, aux;
		for (int i = 1; i < numeros.size(); i++)
		{
			j = i - 1;
			aux = numeros[i];
			while (j >= 0 && (aux < numeros[j]))
			{
				numeros[j+1] = numeros[j];
				j--;
				cont++;
			}
			numeros[j+1] = aux;
		}
		cout << "Numeros de desplazamientos: " << cont << endl;
	}
	static void merge(vector<int>& numeros)
	{
		_merge(numeros, 0, numeros.size() - 1);
	}
};
int main()
{
	srand(time(NULL));
	vector<int>numeros;
	//for (int i = 0; i < n; i++)
	//{
	//	numeros.push_back(rand() % n * 100 + 1);
	//}
	numeros.push_back(5);
	numeros.push_back(3);
	numeros.push_back(6);
	numeros.push_back(1);
	numeros.push_back(9);
	numeros.push_back(2);
	
	print_vec(numeros);
	/*if (busqueda::lineal(numeros, -5))
		cout << "Encontrado" << endl;
	else
		cout << "No encontrado" << endl;
	//ordenamiento::merge(numeros);*/
	cout << "_________Ordenamiento______________" << endl;
	ordenamiento::insercion(numeros);
	print_vec(numeros);
	cout << "Termino el ordenamiento" << endl;
	system("pause>=0");
	return 0;
}
```
